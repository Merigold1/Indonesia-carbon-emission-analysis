```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
data_final= pd.read_csv('data_gabungan.csv')

print(data_final.columns)
```

```python
display(data_final)
```

```python
features = ['Jumlah Total kendaraan', 'Jumlah Industri Besar', 'Emisi Karbon (MtCO2e)']
data_fixed = data_final[features]

```

```python
scaler = StandardScaler()
data_fixed = scaler.fit_transform(data_fixed)
data_fixed
```

```python
# Menentukan jumlah cluster yang optimal dengan Elbow Method
inertia = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_fixed)
    inertia.append(kmeans.inertia_)

# Plot Elbow Method
fig = plt.subplots(figsize=(8, 4))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=2)
data_final['Cluster'] = kmeans.fit_predict(data_fixed)
```

```python
data_final
```

```python
jumlah_cluster = data_final['Cluster'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
jumlah_cluster.plot(kind='bar', color=['orange', 'red'])
plt.xlabel('Cluster')
plt.ylabel('Jumlah Data')
plt.title('Jumlah Data di Setiap Cluster')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

```python
# Menampilkan provinsi yang masuk ke dalam cluster 0
cluster_0 = data_final.loc[data_final['Cluster'] == 0, 'Provinsi']
print("Provinsi yang masuk ke dalam Cluster 0:")
print(cluster_0)

# Menampilkan provinsi yang masuk ke dalam cluster 1
cluster_1 = data_final.loc[data_final['Cluster'] == 1, 'Provinsi']
print("\nProvinsi yang masuk ke dalam Cluster 1:")
print(cluster_1)
```

```python
centroids = kmeans.cluster_centers_


print("Centroid dari setiap cluster:")
for i, centroid in enumerate(centroids):
    print(f"Cluster {i}: {centroid}")
```

```python
# Membuat scatter plot dengan dua kelompok data yang berbeda
plt.scatter(data_final[data_final['Cluster'] == 0]['Jumlah Total kendaraan'], data_final[data_final['Cluster'] == 0]['Emisi Karbon (MtCO2e)'], c='purple', label='Cluster 0')
plt.scatter(data_final[data_final['Cluster'] == 1]['Jumlah Total kendaraan'], data_final[data_final['Cluster'] == 1]['Emisi Karbon (MtCO2e)'], c='yellow', label='Cluster 1')

# Menambahkan keterangan
plt.xlabel('Jumlah Kendaraan')
plt.ylabel('Emisi Karbon')
plt.title('Scatter Plot Jumlah Kendaraan vs Emisi Karbon')
plt.legend(title='Cluster', loc='lower right', fontsize='small')

# Menampilkan plot
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Menghitung Silhouette Score
silhouette_avg = silhouette_score(data_final.drop(columns=['Provinsi', 'Cluster']), data_final['Cluster'])
print("Silhouette Score: ", silhouette_avg)
```

```python
from sklearn.metrics import davies_bouldin_score

# Menghitung Davies-Bouldin Index
db_index = davies_bouldin_score(data_final.drop(columns=['Provinsi', 'Cluster']), data_final['Cluster'])
print("Davies-Bouldin Index: ", db_index)
```

```python
from sklearn.metrics import calinski_harabasz_score

# Menghitung Calinski-Harabasz Index
ch_score = calinski_harabasz_score(data_final.drop(columns=['Provinsi', 'Cluster']), data_final['Cluster'])
print("Calinski-Harabasz Index: ", ch_score)
```

```python
g = sns.FacetGrid(data_final, col="Cluster", col_wrap=3)
g.map_dataframe(sns.scatterplot, x='Jumlah Total kendaraan', y='Emisi Karbon (MtCO2e)')
g.set_axis_labels('Jumlah Kendaraan', 'Emisi Karbon')
g.add_legend()

plt.show()

g = sns.FacetGrid(data_final, col="Cluster", col_wrap=3)
g.map_dataframe(sns.scatterplot, x='Jumlah Industri Besar', y='Emisi Karbon (MtCO2e)')
g.set_axis_labels('Industri Mikro', 'Emisi Karbon')
g.add_legend()

plt.show()
```
